
```python
def merge_inversions_count(lst):
    if len(lst) == 1:
        return (lst,0)
    else:
        (A,i) = merge_inversions_count(lst[:int(len(lst)/2)])
        (B,j) = merge_inversions_count(lst[int(len(lst)/2):])
        (C,k) = merge_and_inversions(A,B)
    return (C,i+j+k)
def merge_and_inversions(A,B):
    i = 0
    j = 0
    num_inversions = 0
    lst = []
    while i < len(A) and j < len(B):
        if A[i]>B[j]:
            lst.append(B[j])
            num_inversions += len(A)-i
            j+=1
        else:
            lst.append(A[i])
            i+=1
    if i == len(A):
        lst.extend(B[j:])
    else:
        lst.extend(A[i:])
    return (lst, num_inversions)
```
