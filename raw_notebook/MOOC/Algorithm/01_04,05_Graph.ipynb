# Dijstras shortest path & Heaps & Balanced BST
## Dijstras Algorithm
### basic
- input: 

 1. directed graph G=(V,E) n=|V|, m=|E|
 2. source vertex __v__
 3. none negtive edge length
- why not BFS -- __blow up the graph__
- the algorithm

```python
X = [s] # list of vertex have been compute
A[s] = 0 # dict of length of s (from v)
B[s] # the route

while V-X is not None:
    for edge(1,2) in cross X and V-X:
        compute the samllest A[1] + len(1,2)
    trans smallest_2 to X
    set A[smallest_2] = A[1] + len(1,2)
    set B[smallest_2] append 1
```

### With Heap
```python
X = [s] # list of vertex have been compute
A[s] = 0 # dict of shortest length of s (from v)
Hp = all vertex length connected to the source v # init

while V-X is not None:
    pop the smallest value (v*, w*) in Hp # O(log(n)) at most n times
    A[w*] = the value
    transfer w* to X
    for every edge(w*, other vertex in V-X):
        if other_v not in Hp:
            Hp[other_v] =  A[w*] +len(w*, other_v) # O(log(n)) at most n times
        else:
            update Hp[other_v] = min( Hp[other_v], A[w*] +len(w*, other_v)) # O(log(n)) at most m times (each edge per time)
```

### Runing time: __O((m + n)log(n)) -> O(m(log(n))__ _since weakly connected_
