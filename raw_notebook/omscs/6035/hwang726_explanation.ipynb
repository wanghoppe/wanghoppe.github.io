# Project 1 - Part 2: Stack Buffer Exploit


## Turn off ASLR - (3 points):
- According to [resouce](https://askubuntu.com/questions/318315/how-can-i-temporarily-disable-aslr-address-space-layout-randomization)
- Set `/proc/sys/kernel/randomize_va_space` interface to 0 to disable ASLR

## Finding the Addresses - (9 points):
- `system`
    - run the program in gdb:
        - `gdb exploit`
        - `run`
    - find address using `p system`.
 
- `/bin/sh`
    - set a breakpoint at main.
    - use `info proc map` to print out memory mapping.
    - choose one address `x` in the previous step and run `find x, +9999999, "/bin/sh"` to find the address of `/bin/sh`.
   
- `exit`
    - similar to `system`, use `p exit`
    - problem: the address of `exit` needs to be put after `system`, but the address is lower than `system`, which can't pass sort function.
    - use `_exit` instead
